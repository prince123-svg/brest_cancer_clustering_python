```python
#import the important libraries
import pandas as pd
```

```python
#load the dataset
old_data=pd.read_csv("D:\\datasets\\intershala\\breast canser\\breast-cancer-wisconsin\\data.csv")
print(old_data)
```

```python
#checking the null values in the dataset
print(old_data.isnull().sum())
```

```python
#shape of the data
print(old_data.shape)
```

```python
#cleaning the dataset 
old_data=old_data.drop(['id'],axis=1)
old_data
```

```python
data=old_data.drop(['diagnosis'],axis=1)
```

```python
old_data
```

```python
data
```

```python
#using PCA(pricipal components )
#Features scaling the data
from sklearn.preprocessing import StandardScaler
scl=StandardScaler()
data=scl.fit_transform(data)
```

```python
#import PCA 

from sklearn.decomposition import PCA
pca=PCA(n_components=None)
data=pca.fit_transform(data)
```

```python
#calculate eigenvalues
x=pca.explained_variance_ratio_
```

```python
#plot the eigenvalues 
import matplotlib.pyplot as plt
plt.plot(x)
plt.show()
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=2, init='k-means++', n_init=10, max_iter=300, tol=0.0001, precompute_distances='auto', verbose=0,
          random_state=None, copy_x=True, n_jobs=1, algorithm='auto')
ky=km.fit_predict(data)
```

```python
old_data['result']=ky
```

```python
old_data['diagnosis']=old_data['diagnosis'].map({'M':1,'B':0})
```

```python
old_data
```

```python
#check the confusion matrix 
from sklearn.metrics import confusion_matrix
confusion_matrix(old_data['diagnosis'],old_data['result'])
```

```python
#check the accuracy 
from sklearn.metrics import accuracy_score
accuracy_score(old_data['diagnosis'],old_data['result'])
```

```python
#classification_score
from sklearn.metrics import classification_report
print(classification_report(old_data['diagnosis'],old_data['result']))
```
